Consider a rope with a knot at each end; these knots mark the head and the tail of the rope. If the head moves far enough away from the tail, the tail is pulled toward the head.

Due to nebulous reasoning involving Planck lengths, you should be able to model the positions of the knots on a two-dimensional grid. Then, by following a hypothetical series of motions (your puzzle input) for the head, you can determine how the tail will move.

Simulate your complete hypothetical series of motions. How many positions does the tail of the rope visit at least once?


```python

with open('data.txt', 'r') as input:
    lines = input.readlines()

tail_visited = {((0, 0))}
head_pos = [0, 0]
tail_pos = [0, 0]

for line_nr, line in enumerate(lines):
    direction, steps = line.split(' ')
    for step in range(int(steps)):
        if direction == 'R':
            head_pos[1] += 1
        if direction == 'U':
            head_pos[0] -= 1
        if direction == 'L':
            head_pos[1] -= 1
        if direction == 'D':
            head_pos[0] += 1
        # decide tail position
        row_diff = head_pos[0] - tail_pos[0]
        col_diff = head_pos[1] - tail_pos[1]
        if abs(row_diff) not in (0, 1) or abs(col_diff) not in (0, 1):
            # move tail
            if col_diff == 0:
                if row_diff > 0:
                    tail_pos[0] += 1
                else:
                    tail_pos[0] -= 1
            elif row_diff == 0:
                if col_diff > 0:
                    tail_pos[1] += 1
                else:
                    tail_pos[1] -= 1
            elif row_diff == 1 and col_diff ==1:
                pass

            elif row_diff > 1:
                tail_pos[0] = head_pos[0] - 1
                tail_pos[1] = head_pos[1]

            elif row_diff < -1:
                tail_pos[0] = head_pos[0] + 1
                tail_pos[1] = head_pos[1]
            
            elif col_diff > 1:
                tail_pos[1] = head_pos[1] - 1
                tail_pos[0] = head_pos[0]

            elif col_diff < -1:
                tail_pos[1] = head_pos[1] + 1
                tail_pos[0] = head_pos[0]

        tail_visited.add(tuple(tail_pos))

print(f'Number of positions that tail visited trees: {len(tail_visited):,}')
```
